```python
import os
import zipfile
import shutil
import random

# === Step 1: Extract the ZIP ===
zip_path = '/content/archive.zip'
extract_path = '/content/dataset'

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# === Step 2: Define Paths ===
source_base = '/content/dataset'
destination_base = '/content/my_10_class_dataset'
splits = ['train', 'test', 'valid']  # use this if folder is named 'valid'

# === Step 3: Pick 10 Random Classes from 'train' folder ===
train_classes_path = os.path.join(source_base, 'train')
all_classes = os.listdir(train_classes_path)
random.seed(42)  # for reproducibility
selected_classes = random.sample(all_classes, 10)

print(" Selected Classes:")
for c in selected_classes:
    print("-", c)
```

```python
# === Step 4: Copy Only Those Classes ===
for split in splits:
    src_split_path = os.path.join(source_base, split)
    dst_split_path = os.path.join(destination_base, split)
    os.makedirs(dst_split_path, exist_ok=True)

    for class_name in selected_classes:
        src_class_dir = os.path.join(src_split_path, class_name)
        dst_class_dir = os.path.join(dst_split_path, class_name)
        if os.path.exists(src_class_dir):
            shutil.copytree(src_class_dir, dst_class_dir)
```

```python
import os

def print_directory_tree(path, indent=""):
    for item in sorted(os.listdir(path)):
        item_path = os.path.join(path, item)
        if os.path.isdir(item_path):
            print(f"{indent} {item}/")
            print_directory_tree(item_path, indent + "    ")
        else:
            print(f"{indent}  {item}")

print("Dataset Directory Structure:\n")
print_directory_tree('/content/my_10_class_dataset')
```

```python
import os

base_path = '/content/my_10_class_dataset'
splits = ['train', 'test', 'valid']

for split in splits:
    split_path = os.path.join(base_path, split)
    total_images = 0
    print(f"\n Split: {split.upper()}")
    print("-" * 40)

    for class_name in sorted(os.listdir(split_path)):
        class_dir = os.path.join(split_path, class_name)
        if os.path.isdir(class_dir):
            num_images = len([
                fname for fname in os.listdir(class_dir)
                if fname.lower().endswith(('.jpg', '.jpeg', '.png'))
            ])
            total_images += num_images
            print(f"{class_name}: {num_images} images")

    print(f" Total images in {split.upper()}: {total_images}")
```

```python
import os
import random
import shutil
from collections import defaultdict
import matplotlib.pyplot as plt

# Set path to TRAIN data
train_dir = '/content/my_10_class_dataset/train'

# Step 1: Count images in each class folder
class_counts = {}
for class_name in os.listdir(train_dir):
    class_path = os.path.join(train_dir, class_name)
    if os.path.isdir(class_path):
        class_counts[class_name] = len(os.listdir(class_path))

# Step 2: Visualize original class distribution
plt.figure(figsize=(10, 5))
plt.bar(class_counts.keys(), class_counts.values())
plt.xticks(rotation=45, ha='right')
plt.ylabel("Image Count")
plt.title("Original Class Distribution (TRAIN)")
plt.tight_layout()
plt.show()

# Step 3: Choose minimum class size for undersampling (POPINJAY has 121 images)
min_samples = min(class_counts.values())

# Step 4: Create a new folder to store balanced dataset
balanced_dir = '/content/my_10_class_dataset/trainB'
os.makedirs(balanced_dir, exist_ok=True)

# Step 5: Undersample all classes to min_samples
for class_name in os.listdir(train_dir):
    class_path = os.path.join(train_dir, class_name)
    if os.path.isdir(class_path):
        images = os.listdir(class_path)
        selected_images = random.sample(images, min_samples)  # random undersampling

        # Create new class folder in balanced dir
        new_class_path = os.path.join(balanced_dir, class_name)
        os.makedirs(new_class_path, exist_ok=True)

        # Copy selected images to the new balanced dataset folder
        for img_name in selected_images:
            src = os.path.join(class_path, img_name)
            dst = os.path.join(new_class_path, img_name)
            shutil.copy(src, dst)

# Step 6: Count new balanced classes
balanced_counts = {}
for class_name in os.listdir(balanced_dir):
    class_path = os.path.join(balanced_dir, class_name)
    balanced_counts[class_name] = len(os.listdir(class_path))

# Step 7: Visualize balanced dataset
plt.figure(figsize=(10, 5))
plt.bar(balanced_counts.keys(), balanced_counts.values(), color='green')
plt.xticks(rotation=45, ha='right')
plt.ylabel("Image Count")
plt.title("Balanced Class Distribution (TRAIN)")
plt.tight_layout()
plt.show()

# Summary print
print(" Original image count per class:", class_counts)
print(" Balanced image count per class:", balanced_counts)
```

```python
# Core PyTorch
import torch
import torch.nn as nn
import torch.optim as optim

# TorchVision for image datasets and transforms
from torchvision import datasets, transforms

# DataLoader for batching data
from torch.utils.data import DataLoader

# For evaluation and metrics
from sklearn.metrics import classification_report, confusion_matrix

# Plotting and utilities
import numpy as np
import matplotlib.pyplot as plt
import os
import random
```

```python
# For training
train_transforms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ToTensor(),
    transforms.Normalize([0.5], [0.5])
])

# For validation/test (no augmentation)
valid_test_transforms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.5], [0.5])
])
```

```python
# Define the directory paths
train_dir = '/content/my_10_class_dataset/trainB'
valid_dir = '/content/my_10_class_dataset/valid'
test_dir = '/content/my_10_class_dataset/test'

# Load datasets using ImageFolder
train_dataset = datasets.ImageFolder(train_dir, transform=train_transforms)
valid_dataset = datasets.ImageFolder(valid_dir, transform=valid_test_transforms)
test_dataset = datasets.ImageFolder(test_dir, transform=valid_test_transforms)

# Create DataLoaders
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
valid_loader = DataLoader(valid_dataset, batch_size=16, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)

```

```python
class SimpleCNN(nn.Module):
    def __init__(self, num_classes=10):
        super(SimpleCNN, self).__init__()
        self.network = nn.Sequential(
            nn.Conv2d(3, 16, 3, padding=1),
            nn.BatchNorm2d(16),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(16, 32, 3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(32, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Flatten(),
            nn.Dropout(0.5),
            nn.Linear(64 * 28 * 28, 128),
            nn.ReLU(),
            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        return self.network(x)

device = torch.device('cpu')

# Now define and move the model to CPU
model = SimpleCNN().to(device)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
for epoch in range(10):
    model.train()
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

    # Validation phase
    model.eval()
    with torch.no_grad():
        correct = 0
        total = 0
        for inputs, labels in valid_loader:
            outputs = model(inputs)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    print(f"Epoch {epoch+1}: Validation Accuracy = {100 * correct / total:.2f}%")
```

```python
model.eval()

all_preds = []
all_labels = []

with torch.no_grad():
    for inputs, labels in test_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        outputs = model(inputs)
        _, predicted = torch.max(outputs, 1)
        all_preds.extend(predicted.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

all_preds = np.array(all_preds)
all_labels = np.array(all_labels)

# Classification report
print("Classification Report:\n")
print(classification_report(all_labels, all_preds, target_names=test_dataset.classes))

# Confusion matrix
print("\nConfusion Matrix:\n")
print(confusion_matrix(all_labels, all_preds))

# Accuracy
correct = (all_preds == all_labels).sum()
total = len(all_labels)
accuracy = 100 * correct / total
print(f"\n Test Accuracy: {accuracy:.2f}%")

```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix

# Generate the confusion matrix
cm = confusion_matrix(all_labels, all_preds)
classes = test_dataset.classes

# Plot using only matplotlib
fig, ax = plt.subplots(figsize=(10, 8))
cax = ax.matshow(cm, cmap='Greens')
fig.colorbar(cax)

# Set axis labels and ticks
ax.set_xticklabels([''] + classes, rotation=45, ha='left')
ax.set_yticklabels([''] + classes)
ax.set_xlabel('Predicted')
ax.set_ylabel('True')
ax.set_title('Confusion Matrix - Butterfly & Moth Classification')

# Annotati values
for (i, j), val in np.ndenumerate(cm):
    ax.text(j, i, str(val), ha='center', va='center', color='black')

plt.tight_layout()
plt.show()
```

```python
import torch
import torch.nn as nn
from torchvision import transforms
from PIL import Image

# Define the model architecture (same as training)
class SimpleCNN(nn.Module):
    def __init__(self, num_classes=10):
        super(SimpleCNN, self).__init__()
        self.network = nn.Sequential(
            nn.Conv2d(3, 16, 3, padding=1),
            nn.BatchNorm2d(16),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(16, 32, 3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(32, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Flatten(),
            nn.Dropout(0.5),
            nn.Linear(64 * 28 * 28, 128),
            nn.ReLU(),
            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        return self.network(x)

# Load class names
class_names = ['BANDED TIGER MOTH', 'CHALK HILL BLUE', 'EMPEROR GUM MOTH',
               'GLITTERING SAPPHIRE', 'GREEN HAIRSTREAK', 'MOURNING CLOAK',
               'POPINJAY', 'PURPLISH COPPER', 'RED CRACKER', 'ROSY MAPLE MOTH']

# Load model and weights
model = SimpleCNN(num_classes=10)
model.load_state_dict(torch.load('butterfly_model.pth', map_location='cpu'))
model.eval()
```

```python
# Define image preprocessing (same as validation/test)
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.5], [0.5])
])

def predict_image(img):
    img = img.convert("RGB")
    img_tensor = transform(img).unsqueeze(0)
    with torch.no_grad():
        output = model(img_tensor)
        _, predicted = torch.max(output, 1)
    return class_names[predicted.item()]
```
